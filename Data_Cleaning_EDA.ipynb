##### Load scraped yelp reviews

```python
# yelp_raw_data = pd.read_csv('yelp_reviews.csv', index_col=0)
yelp_raw_data = pd.read_csv('yelp_reviews.csv')
yelp_raw_data.head()
```

```python
location_, price_, review_count, reviews_, restaurant_name = [], [], [], [], []
```

##### Function to parse raw data for locations into dataframe

```python
def parse_location(df, new_df):
    df_split = ','.join([str(elem) for elem in df]).split(',')
    new_df = [','.join(df_split[i:i+2]) for i in range(0, len(df_split), 2)]
    return new_df
```

##### Parsing location data

```python
location_ = parse_location(yelp_raw_data.City, location_)
location_
```

##### Function to parse raw data into dataframe

```python
# Function to parse dataframe based on single comma split followed by unlisting nested lists
def parse_dataframe(df, new_df):
    df = list(map(lambda x: x.split(','), df))
    for i in df:
        for j in i:
            new_df.append(j)
    return new_df
```

##### Parsing restaurant price data

```python
parse_dataframe(yelp_raw_data.Dollar_sign, price_)
price_
```

##### Parsing restaurant review counts

```python
parse_dataframe(yelp_raw_data.Nbr_reviews, review_count)
review_count
```

##### Function to remove text string from ratings

```python
def numeric_rating(df):
    return df.replace(' star rating', '', regex = True)
```

##### Removing string from ratings data

```python
ratings_0 = numeric_rating(yelp_raw_data.Rating)
```

##### Parsing ratings data

```python
parse_dataframe(ratings_0, reviews_)
reviews_
```

##### Function to parse raw data for restaurants names into dataframe

```python
# Function to parse dataframe based on single comma split followed by unlisting nested lists
def parse_restname(df, new_df):
    df = list(map(lambda x: x.replace(', ', ' ').split(','), df))
    for i in df:
        for j in i:
            new_df.append(j)
    return new_df
```

##### Parsing the restaurant names

```python
parse_restname(yelp_raw_data.RestName, restaurant_name)
restaurant_name[:5]
```

##### Combine cleaned dataset into single dataframe

```python
yelp_reviews = pd.DataFrame(list(zip(location_, price_, review_count, reviews_, restaurant_name)), columns = ['City', \
                'Price_Range', 'Number_of_Reviews', 'Rating', 'Restaurant_Name'])
yelp_reviews.sample(4)
```

```python
yelp_reviews.City.unique()
```

##### Convert Rating and Number of Rewiews columns to int/float type

```python
yelp_reviews['Rating'] = yelp_reviews['Rating'].astype(float)
yelp_reviews['Number_of_Reviews'] = yelp_reviews['Number_of_Reviews'].astype(int)

yelp_reviews.head()
```

##### Create column to show the description of the price range for each restaurant

```python
# $= under $10. $$=11-30. $$$=31-60. $$$$= over $61.

yelp_reviews.loc[yelp_reviews['Price_Range'] == '$', "Price_Description"] = "under $10"
yelp_reviews.loc[yelp_reviews['Price_Range'] == '$$', "Price_Description"] = "\$11 - $30"
yelp_reviews.loc[yelp_reviews['Price_Range'] == '$$$', "Price_Description"] = "\$31 - $60"
yelp_reviews.loc[yelp_reviews['Price_Range'] == '$$$$', "Price_Description"] = "over $61"
yelp_reviews.sample(10)
```

##### Create column to show the category of each restaurant based on price

```python

yelp_reviews.loc[yelp_reviews['Price_Range'] == '$', "Category"] = "Low-end"
yelp_reviews.loc[yelp_reviews['Price_Range'] == '$$', "Category"] = "Middle-end"
yelp_reviews.loc[((yelp_reviews['Price_Range'] == ('$$$')) | (yelp_reviews['Price_Range'] == ('$$$$'))), "Category"] = "High-end"
yelp_reviews.sample(10)
```

##### Create new column 'Price_Value' as a numeric value for price range; i.e. $$ refers to two price levels

```python
yelp_reviews['Price_Value'] = yelp_reviews.apply(lambda row: len(row.Price_Range),axis=1)
```

```python
yelp_reviews.sample(10)
```
